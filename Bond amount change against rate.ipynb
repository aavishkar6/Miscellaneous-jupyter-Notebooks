```python
parValue = 10000
rate = 15
n = 3
bp = []

for i in range(1,100):
    value = annuity(parValue, i/100, n) + present_value( parValue, i/100, n)
    bp.append(value)
```

```python
def annuity(par, i , n):
    coef = ( 1 + i)** n
    total_coef = 1 - 1/coef
    return (par/i) * total_coef
```

```python
def present_value(par, i, n):
    coef = 1 + i
    coef = coef ** n
    return (par/coef)
```
